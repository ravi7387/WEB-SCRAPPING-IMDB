```python

import requests as r
from bs4 import BeautifulSoup as bs
import csv
import pandas as pd

movie_name = []

year = []

rating = []

genre = []

run_time = []

dirc = []

director = []

for i in range(1,250,50):

  url = "https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start={}&ref_=adv_nxt".format(i)

  source = r.get(url).content

  soup = bs(source,"lxml")

  tag = soup.find_all("h3",class_="lister-item-header")

  tag1 = soup.find_all("span",class_="lister-item-year text-muted unbold")

  tag2 = soup.find_all("div",class_="inline-block ratings-imdb-rating")

  tag3 = soup.find_all("span",class_="genre")

  tag4 = soup.find_all("span",class_="runtime")

  tag5 = soup.find_all("div",class_="lister-item-content")


  for i,j,k,l,m in zip(tag,tag1,tag2,tag3,tag4):
    movie_name.append(i.find("a").text)
    year.append(j.text.strip("()"))
    rating.append(k.find("strong").text)
    genre.append(l.text.strip())
    run_time.append(m.text)

  tag5 = soup.find_all("div",class_="lister-item-content")
  for n in tag5:
    dirc.append(n.find_all("p")[2].text)
  for g in dirc:
    director.append(g.split("\n")[2].strip())

v = list(zip(movie_name,year,rating,genre,run_time,director))

a = pd.DataFrame(v)

a.columns = ["MOVIE_NAME","RELEASED YEAR","RATING","genre","RUN_TIME","DIRECTOR"]

a.to_csv('movies.csv',index=None)

a



```
